```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx"
df = pd.read_excel(url, sheet_name='Canada by Citizenship',skiprows=range(20), skipfooter=2)
df.tail()
```

```python
# to remove select colums
df.drop(columns=['Type','Coverage','DEV','AREA','REG'],inplace=True)
```

```python
df.rename(columns={'OdName':'Country','AreaName':'Continent'}, inplace=True)
```

```python
df.set_index('Country',inplace=True)
# df.reset_index(inplace=True)
```

```python
years = list(range(1980,2014))
years
```

```python
df.loc['India',years].plot()
plt.title('India Immigration from 1980 to 2013')
plt.xlabel(years)
plt.ylabel('Immigration Count')
plt.show()
```

```python
df.loc['China',years].plot(figsize=(10,5))
plt.title('China Immigration from 1980 to 2013')
plt.xlabel(years)
plt.ylabel('Immigration Count', size=20)
plt.show()
```

```python
# transposing data
df.loc[['China','India'],years]
```

```python
df[df['Continent']== 'Asia'][years].sort_values(2013, ascending=False).head(10).T.plot(figsize=(14,6))
plt.xticks(size=12, rotation=30)
plt.yticks(size=12)
plt.show()
```

```python
 # [[2011,2012,2013]].T.plot(kind='bar')
df
```

```python
df[years].sum().plot(kind='bar',figsize=(15,6))
```

```python
years = list(range(2010, 2011))
df[df["Continent"] == "Europe"][years].sort_values (2010, ascending=False).head(10).plot(kind='bar')
plt.title('Europe (Top 10 Countries) Immigration counts for the year 2010', size = 15)
plt.xlabel('Countries', size = 12)
plt.ylabel('Immigration Count', size = 12)
plt.xticks(size=10, rotation=-90)
plt.yticks(size=8)
plt.show()
```
